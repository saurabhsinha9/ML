#Binary Classification

```python
import pandas as pd
df=pd.DataFrame(
    [(22,0),(25,0),(47,1),(52,0),(46,1),(56,1),(55,0),(60,1),(62,1),(61,1),(18,0),(28,0),(27,0),(29,0),(49,1),(55,1),(25,1),(58,1),(19,0),(18,0),(21,0),(26,0),(40,1),(45,1),(50,1),(54,1),(23,0)]
    ,columns=['age','insurance_purchased']           
               )
df.head(5)
```

```python
import matplotlib.pyplot as plt
plt.xlabel("age")
plt.ylabel("isurance purchased")
plt.scatter(df.age,df.insurance_purchased)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['age']],df.insurance_purchased,test_size=0.1)
```

```python
from sklearn import linear_model
lr=linear_model.LogisticRegression()
lr.fit(X_train,y_train)
y_test
```

```python
lr.predict(X_test)
```

```python
lr.score(X_train,y_train)
```

```python
lr.get_params()
```

```python
lr.predict_proba(X_test)
```

```python
plt.xlabel("age")
plt.ylabel("isurance purchased")
plt.scatter(X_train,y_train, label='training')
plt.plot(X_test,lr.predict(X_test), label='test')
plt.legend()
```
